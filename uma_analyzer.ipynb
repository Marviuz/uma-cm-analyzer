```
from uma_analyzer.colab import quick_report

# Use SHEET_URL or csv_path
report = quick_report(SHEET_URL)

# Strategy Performance
print("=== Strategy Performance ===")
for s in report['strategy_stats']:
    print(f"{s.strategy.name}: Count={s.count}, Avg={s.avg_rank:.2f}, Top3={s.top3_rate*100:.1f}%")

# Stat Correlations
print("\n=== Stat Correlations ===")
for c in report['correlations']:
    print(f"{c.stat}: {c.correlation:.3f}")

# Top Skills
print("\n=== Top 5 Skills ===")
for skill in report['skill_impacts'][:5]:
    print(f"{skill.name}: SIS={skill.sis:.3f}")

# Success Envelope
print("\n=== Success Envelope ===")
e = report['envelope']
print(f"Min Speed for Top3: {e.min_speed_top3}")
print(f"Avg Stamina: {e.avg_stamina_finishers:.0f} ± {e.stamina_std:.0f}")
```
